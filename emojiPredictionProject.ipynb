## Step 3   To convert each sentence into embeddings 
  - Download Glove6B50d.txt

```python
f = open("glove.6B.50d.txt" , encoding='utf8')
```

```python
embeddings_index = {} 

for line in f : 
    values = line.split() 
    word = values[0] 
    coefs = np.asarray(values[1:] , dtype ='float')
    # print(word , coefs)
    embeddings_index[word] = coefs
f.close()
```

```python
#embeddings_index 

## We have hence built an entire dictionary of 6 billion words and their corresponding vectors
```

```python
embeddings_index['eat']
```

```python
embeddings_index['eat'].shape
```

```python
emd_dim = embeddings_index['eat'].shape[0]
print(emd_dim)
```

Now here we have seen how to read and load ur glove vecotrs and make ur custom dictionary 

## Step 4 : Converting sentences into vectors ; this would create our embedding layer output

```python
def embedding_output(X)  :
    maxLen = 10 ## this can be varied acc to the length of ur sentences 
    
    embedding_matrix_output = np.zeros((X.shape[0],maxLen, emd_dim)) ## 3D 
    for ix in range(X.shape[0]) : 
        X[ix] =X[ix].split() ## to get words out of each sentence          
        for jx in range(len(X[ix])) : 
            embedding_matrix_output[ix][jx] = embeddings_index[X[ix][jx].lower()]
            
    return embedding_matrix_output
```

```python
print(X_train[0])
print(X_test[0])
X_train.shape
```

```python
embedding_matrix_train = embedding_output(X_train)
embedding_matrix_test = embedding_output(X_test)
```

```python
from keras.utils import to_categorical
```

```python
Y_train = to_categorical(Y_train, num_classes=5)
Y_test = to_categorical(Y_test, num_classes=5)
```

## Step 5 - Define the RNN/LSTM model

```python
from keras.models import Sequential 
from keras.layers import LSTM , Dense ,Dropout
```

```python
model = Sequential()
model.add(LSTM(64, input_shape= (10,50)))   ## a LSTM cell added 
## each sentence with 10words and each word with a max dimension of 50---- 
## hidden state should have 64 dimensions
model.add(Dropout(0.5))
model.add(Dense(5 , activation = 'softmax')) 
model.compile(loss= 'categorical_crossentropy' ,optimizer='adam', metrics = ['accuracy'])

model.summary()
```

```python
from keras.callbacks import ModelCheckpoint , EarlyStopping

checkpoints= ModelCheckpoint("best_model.h5"  ,verbos =True , save_best_only=True , monitor='val_loss')
EarlyStopping(monitor = 'val_acc',patience=10)
hist = model.fit(embedding_matrix_train ,Y_train , epochs = 100  ,batch_size=64 , shuffle =True , validation_split=0.2)
```

```python
pred = model.predict_classes(embedding_matrix_test)
```

```python
print(pred)
```

```python
model.evaluate(embedding_matrix_test,Y_test)
```

## building stacked LSTM 

```python
model = Sequential()
model.add(LSTM(64, input_shape= (10,50) , return_sequences = True))  ## the lower level will return sequences to the upper stacked LSTM as input   
model.add(Dropout(0.5))
model.add(LSTM(64,return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(5 , activation = 'softmax')) 
model.compile(loss= 'categorical_crossentropy' ,optimizer='adam', metrics = ['accuracy'])

model.summary()
```
